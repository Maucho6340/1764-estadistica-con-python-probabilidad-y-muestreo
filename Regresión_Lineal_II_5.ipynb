*Test de significancia conjunta de los parámetros*<br>
Prob (F-statistic) <= 0.05 (OK)

*Test de significancia individual de los parámetros*<br>
P>|t| <= 0.05 (OK)

**La prueba F**: <br>
Evalúa la significancia conjunta de varios parámetros dentro de un modelo estadístico. Es una herramienta clave para determinar si un grupo de variables explicativas, tomadas en conjunto, contribuye de manera significativa a explicar la variable dependiente.

**¿Cómo funciona en statsmodels?**<br>
Cuando usas .summary() sobre un modelo de regresión, la prueba F:
- Compara el modelo completo contra un modelo reducido (sin variables explicativas).
- Calcula el estadístico F con su respectivo p-value, que indica si hay evidencia de que el modelo tiene poder explicativo.
- Si el p-value es bajo (ej. < 0.05), puedes concluir que al menos una variable explicativa tiene un efecto sobre la variable dependiente.<br>

**La prueba t**:<br>
En modelos de regresión (como los que estimamos con statsmodels) se utiliza precisamente para evaluar si cada coeficiente individual es significativamente distinto de cero. En otras palabras, analiza si cada variable explicativa tiene un efecto significativo sobre la variable dependiente. <br>

**¿Cómo funciona?**<br>
- Para cada parámetro estimado (cada variable), se calcula un estadístico t:
Donde:
- \( \hat{\beta}_i \) es el valor estimado del coeficiente
- \( \text{SE}(\hat{\beta}_i) \) es su error estándar
- Luego se verifica su p-value, y si es menor al umbral (por ejemplo 0.05), se considera estadísticamente significativo.
En el resumen del modelo (.summary()), verás una columna P>|t| que muestra estos p-values para cada variable.

### Coeficiente de Determinación - R²

El coeficiente de determinación (R²) es una medida de resumen que indica cuánto se ajusta la recta de regresión a los datos. Es un valor entre 0 y 1.


$$R^2(y, \hat{y}) = 1 - \frac {\sum_{i=0}^{n-1}(y_i-\hat{y}_i)^2}{\sum_{i=0}^{n-1}(y_i-\bar{y}_i)^2}$$

```python
print('R²= {}'.format(modelo.score(X_train,y_train)))
```

Obtenemos el mismo R2 para los datos de entrenamiento y prueba
