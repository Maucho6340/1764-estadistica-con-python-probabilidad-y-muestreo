## Importando la biblioteca statsmodels

https://www.statsmodels.org/stable/index.html

A diferencia de scikit-learn, debemos indicar específicamente que queremos la constante modelo Beta1, de o contrario, la función statsmodels no incluirá la constante. Para incluir la constante modelo, debemos incluirla en el DataFrame del conjunto de entrenamiento formado por las variables explicativas X_train.

```python
# Importar biblioteca
import statsmodels.api as sm
```

## Estimando el modelo con statsmodels

```python
# Crea nuevo data frame que contiene la constante Beta
X_train_con_constante = sm.add_constant(X_train)
```

```python
X_train_con_constante
```

Con esto ya hemos uncluido la constante en el DF y podemos usar statsmodels, recordando que scikit-learn no necesitamos hacer este procedimiento debido a que por defecto la función scikit_learn incluye la constante modelo.<br>

**¿Que pasa si no incluimos la constante?**<br>
El modelo de regresión puede hacerse sin la constante, no hay obligación de incluirla, pero la aproximación del modelo con constante, aegura que los residuos, que son aproximación de ui (al final de la fórmula), tenga un pomedio de cero, pues uno de los presupuestos teóricos de regresión, es que los residuos tengan valor cero. Al no asegurar esto, vamos a tener relaciones sesgadas.

```python
# Creación del modelo de regresión lineal
# Nombre: modelo_statsmodels

# sm.OLS: Llamamos a la clase OLS (Ordinary Least Squares, o mínimos cuadrados ordinarios)
#         proveniente de la biblioteca statsmodels, que permite estimar modelos de regresión lineal.

# Parámetros esperados por sm.OLS:
# - y_train: Variable dependiente del conjunto de entrenamiento.
# - X_train_con_constante: DataFrame con variables explicativas y la constante añadida manualmente.
# - hasconst=True: Indicamos que la constante ya está incluida en X_train_con_constante,
#                  lo que evita que statsmodels la añada automáticamente.

# .fit(): Método que ajusta el modelo y estima los coeficientes mediante mínimos cuadrados.

modelo_statsmodels = sm.OLS(y_train, X_train_con_constante, hasconst=True).fit()
```

### Coeficiente de Determinación - R²

El coeficiente de determinación (R²) es una medida de resumen que indica cuánto se ajusta la recta de regresión a los datos. Es un valor entre 0 y 1.


$$R^2(y, \hat{y}) = 1 - \frac {\sum_{i=0}^{n-1}(y_i-\hat{y}_i)^2}{\sum_{i=0}^{n-1}(y_i-\bar{y}_i)^2}$$
